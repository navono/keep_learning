```python
import jmespath


def get_center_shift_box(box):
    box_left = box[0][0]
    box_up = box[0][1]
    box_right = box[2][0]
    box_down = box[2][1]

    down_shift = 100
    cx = (box_left + box_right) / 2
    cy = ((box_up + box_down) / 2) + down_shift

    thresh_hold = 2
    return [
        [cx - thresh_hold, cy + thresh_hold],
        [cx - thresh_hold, cy - thresh_hold],
        [cx + thresh_hold, cy - thresh_hold],
        [cx + thresh_hold, cy + thresh_hold]
    ]


def is_box_contained(src_box, target_box):
    """
    判断 target_box 是否被 src_box 包含
    """

    src_box_left = src_box[0][0]
    src_box_up = src_box[0][1]
    src_box_right = src_box[2][0]
    src_box_down = src_box[2][1]

    return target_box[0][0] > src_box_left and target_box[0][1] > src_box_up and \
        target_box[2][0] < src_box_right and target_box[2][1] < src_box_down


row_json = [
    {
        "box": [
            [
                5,
                52
            ],
            [
                24,
                52
            ],
            [
                24,
                97
            ],
            [
                5,
                97
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.0,
        "text": ""
    },
    {
        "box": [
            [
                5124,
                99
            ],
            [
                5299,
                105
            ],
            [
                5296,
                175
            ],
            [
                5121,
                169
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.9469092488288879,
        "text": "P-106B"
    },
    {
        "box": [
            [
                4987,
                205
            ],
            [
                5427,
                205
            ],
            [
                5427,
                263
            ],
            [
                4987,
                263
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.9997313022613525,
        "text": "D-106B夹套水循环泵"
    },
    {
        "box": [
            [
                6876,
                330
            ],
            [
                6999,
                330
            ],
            [
                6999,
                380
            ],
            [
                6876,
                380
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.9918946027755737,
        "text": "会签专业"
    },
    {
        "box": [
            [
                6885,
                369
            ],
            [
                6975,
                369
            ],
            [
                6975,
                408
            ],
            [
                6885,
                408
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.7828730344772339,
        "text": "DISC."
    },
    {
        "box": [
            [
                6958,
                447
            ],
            [
                6997,
                447
            ],
            [
                6997,
                486
            ],
            [
                6958,
                486
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.535336971282959,
        "text": "Cou"
    },
    {
        "box": [
            [
                1238,
                522
            ],
            [
                1274,
                522
            ],
            [
                1274,
                561
            ],
            [
                1238,
                561
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.9620883464813232,
        "text": "2"
    },
    {
        "box": [
            [
                3568,
                522
            ],
            [
                3603,
                522
            ],
            [
                3603,
                563
            ],
            [
                3568,
                563
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.9881609678268433,
        "text": "5"
    },
    {
        "box": [
            [
                4344,
                522
            ],
            [
                4380,
                522
            ],
            [
                4380,
                566
            ],
            [
                4344,
                566
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.9864794611930847,
        "text": "6"
    },
    {
        "box": [
            [
                5898,
                522
            ],
            [
                5933,
                522
            ],
            [
                5933,
                563
            ],
            [
                5898,
                563
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.6828521490097046,
        "text": "8"
    },
    {
        "box": [
            [
                6671,
                519
            ],
            [
                6710,
                519
            ],
            [
                6710,
                563
            ],
            [
                6671,
                563
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.0,
        "text": ""
    },
    {
        "box": [
            [
                2023,
                530
            ],
            [
                2042,
                530
            ],
            [
                2042,
                544
            ],
            [
                2023,
                544
            ]
        ],
        "cls_label": -1,
        "cls_score": 4.590793898974533e-41,
        "score": 0.0,
        "text": ""
    }
]
target_box = [
    [
        5124,
        99
    ],
    [
        5299,
        105
    ],
    [
        5296,
        175
    ],
    [
        5121,
        169
    ]
]

target_box_center = get_center_shift_box(target_box)
print("target_box_center: ", target_box_center)
result = is_box_contained(src_box, target_box_center)
print("result: ", result)

# 使用 JMESPath 提取所有 box 字段
expression = jmespath.compile('[*].box')
boxes = expression.search(row_json)
print(boxes)
is_contained = any(is_box_contained(box, target_box_center) for box in boxes)
print(is_contained)

expression2 = jmespath.compile('[*].{box: box, text: text}')
boxes_and_texts = expression2.search(row_json)
# print(boxes_and_texts)

matching_text = None
for item in boxes_and_texts:
    print(item['box'])
    if is_box_contained(item['box'], target_box_center):
        matching_text = item['text']
        break
print(matching_text)
```
