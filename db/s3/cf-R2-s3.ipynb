```python
import boto3
from IPython.display import Image, display
```

```python
# 使用 ~/.asw/credentials 配置
# s3 = boto3.client("s3", endpoint_url="http://localhost:8333")

s3 = boto3.client("s3", endpoint_url="https://x.r2.cloudflarestorage.com",
                  aws_access_key_id='x',
                  aws_secret_access_key='x',
                  region_name='auto')

buckets = s3.list_buckets()

# 遍历存储桶列表
for bucket in buckets["Buckets"]:
    print(bucket["Name"])

obj_info = s3.head_object(Bucket="pic", Key="GitHubCopilot.png")
print(obj_info)
```

```python
obj_pic = s3.get_object(Bucket="pic", Key="GitHubCopilot.png")
print(obj_pic)

body = obj_pic["Body"].read()
display(Image(body))
```
